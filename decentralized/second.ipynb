```python
import pandas as pd
from gurobipy import *
import numpy as np
```

```python
data = pd.read_excel('焚化廠資料.xlsx')
pd.set_option('display.max_rows', None)
data.columns=['name', 'city', 'year', 'input_other', 'input_labor', 'input_garbage', 'fixed_input_1', 'fixed_input_2', 'output', 'undesirable_output']
data
```

```python
names = np.unique(data['name'])
grouped_year = data.groupby(data.year)

N = names.size
input_name = ['input_other', 'input_labor', 'input_garbage']
fixed_input_name = ['fixed_input_1', 'fixed_input_2']

record = {}
for plant in names:
    record[plant] = {}

for year in range(2011, 2021):
    for plant in names:
        
        M = Model('NT')
        M.Params.LogToConsole = 0
        DATA = grouped_year.get_group(year)
        DATA_MAIN = DATA[DATA['name'] == plant]

        _lambda = {}
        for i in range(N):
            _lambda[i] = M.addVar(lb = 0, ub = 1, name='lambda%d'%(i))

        theta = M.addVar(lb = 0, name='theta')

        w = {}
        for i in range(N):
            w[i] = M.addVar(lb = 0, ub = 1, name='w%d'%(i))

        M.update()

        M.setObjective(theta * DATA_MAIN['output'], GRB.MAXIMIZE)

        for _input in input_name:
            M.addConstr(quicksum(np.array(DATA[_input])[i] * _lambda[i] for i in range(N)) <= DATA_MAIN[_input])

        for _input in fixed_input_name:
            M.addConstr(quicksum(np.array(DATA[_input])[i] * _lambda[i] for i in range(N)) <= DATA_MAIN[_input])

        M.addConstr(quicksum(np.array(DATA['output'])[i] * _lambda[i] for i in range(N)) - theta * DATA_MAIN['output'] >= 0)
        M.addConstr(quicksum(_lambda[i] for i in range(N)) == 1)


        for _input in fixed_input_name:
            M.addConstr(quicksum(np.array(DATA[_input])[i] * w[i] for i in range(N)) >= DATA_MAIN[_input])

        M.addConstr(quicksum(np.array(DATA['undesirable_output'])[i] * w[i] for i in range(N)) - DATA_MAIN['undesirable_output'] >= 0)
        M.addConstr(quicksum(w[i] for i in range(N)) == 1)

        for _input in fixed_input_name:
            M.addConstr(quicksum(np.array(DATA[_input])[i] * _lambda[i] for i in range(N)) - quicksum(np.array(DATA[_input])[i] * w[i] for i in range(N)) == 0)
        
        M.optimize()
        
        record[plant][year] = {
            'objval': M.objVal,
            'original output': float(DATA_MAIN['output']),
            'theta': [v.x for v in M.getVars() if v.varName in ['theta']][0]
        }
```
