```python
# Sets
Year = list(range(2011, 2021))
K = list(range(24))  # number of DMUs
I_V = [0,1,2]  # variable input amount
I_F = [0,1]  # fixed input amount
J = [0]  # output 
M = [0]  # bad output
years = list(range(1,10))
dmus = list(range(1,25))
```

```python
year = 2020
# Parameters
X_V = [
    list(df[df["年份"] == year]["其他成本"].values),
    list(df[df["年份"] == year]["勞動成本"].values),
    list(df[df["年份"] == year]["焚化處理量"].values)
]
X_F = [
    list(df[df["年份"] == year]["設計處理量"].values),
    list(df[df["年份"] == year]["設計發電量"].values)
]
Y = [list(df[df["年份"] == year]["實際發電量"].values)]
B = [list(df[df["年份"] == year]["二氧化碳當量"].values)]
```

```python
NT = [0 for r in K]
lambda_NT = [[] for r in K]
omega_NT = [[] for r in K]

for r in K:
    md11 = Model('model')
    _lambda = md11.addVars(K, vtype=GRB.CONTINUOUS, name='lambda')
    _omega = md11.addVars(K, vtype=GRB.CONTINUOUS, name='omega')
    _theta = md11.addVar(vtype=GRB.CONTINUOUS, name='theta')  

    md11.setObjective(
        quicksum(_theta * Y[j][r] for j in J), GRB.MAXIMIZE
    )

    md11.addConstrs(
        quicksum(_lambda[k] * X_F[i][k] for k in K) <= X_F[i][r]
        for i in I_F
    )

    md11.addConstrs(
        quicksum(_lambda[k] * X_V[i][k] for k in K) <= X_V[i][r]
        for i in I_V
    )

    md11.addConstrs(
        quicksum(_lambda[k] * Y[j][k] for k in K) >= _theta * Y[j][r]
        for j in J
    )

    md11.addConstr(
        quicksum(_lambda[k] for k in K) == 1
    )

    md11.addConstr(
        quicksum(_omega[k] * X_V[2][k] for k in K) >=  X_V[2][r]
    )

    md11.addConstrs(
        quicksum(_omega[k] * B[m][k] for k in K) <=  B[m][r]
        for m in M
    )

    md11.addConstr(
        quicksum(_omega[k] for k in K) == 1
    )

    md11.addConstr(
        quicksum(_omega[k] * X_V[2][k] for k in K) ==  quicksum(_lambda[k] * X_V[2][k] for k in K)
    )

    md11.params.LogToConsole = 0
    # md11.params.NonConvex = 2
    md11.optimize() 

    print(df.loc[r,"焚化廠名稱"])
    for var in md11.getVars():
        print(f"{var.VarName}, {var.X}")

    lambda_list = md11.getAttr('X', _lambda)
    omega_list = md11.getAttr('X', _omega)
    lambda_NT[r] = lambda_list
    omega_NT[r] = omega_list

    # to load crs
    NT[r] = md11.objVal
```

```python
writer = pd.ExcelWriter('../results/NT.xlsx')

pd.DataFrame(lambda_NT).to_excel(writer, sheet_name='lambda')
pd.DataFrame(omega_NT).to_excel(writer, sheet_name='omega')
writer.save()
```
